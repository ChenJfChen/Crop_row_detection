```python
import cv2

# Input: 60 fps video
# Output: 2 fps video

cap = cv2.VideoCapture(r'C:\Users\khan1\jupyter_test_code\crop-row.mp4')
count = 0

#original video size
height = 1346  #height = frame.shape[0]
width = 1994   #width = frame.shape[1]

#resized video size for before cropping
new_height = 972 
new_width = 1444

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(r'C:\Users\khan1\jupyter_test_code\2fps_crop_row.avi',fourcc, 2.0, (width,height), True)

while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    if ret:
        #cv2.imwrite('frame{:d}.jpg'.format(count), frame)
        out.write(frame)
        count += 30 # i.e. at 60 fps, this advances 0.5 second
        cap.set(1, count) #cap.set(1,frame_no); # Where frame_no is the frame you want
        cv2.imshow("frame",frame)
        print("showing frame: ", count)
        
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()       
out.release()
cv2.destroyAllWindows()
```
